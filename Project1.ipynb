```python
import pandas as pd
```

```python
customer_list = pd.read_excel('.//customer_call List.xlsx')
```

```python
customer_list.info()
```

```python
customer_list.duplicated()
```

```python
customer_list.drop_duplicates(inplace = True)
```

```python
customer_list
```

```python
customer_list.shape
```

```python
customer_list.drop(columns='Not_Useful_Column', inplace=True)
```

```python
customer_list
```

```python
#stripping the first name to make sure there is no whitespace
customer_list['First_Name']=customer_list['First_Name'].str.strip()
```

```python
customer_list
```

```python
customer_list['Last_Name']=customer_list['Last_Name'].str.strip('._/123')# regular expression. there lstrip & rstrip
```

```python
customer_list
```

```python
customer_list['Phone_Number']= customer_list['Phone_Number'].astype(str)
```

```python
customer_list['Phone_Number'] = customer_list['Phone_Number'].str.replace(r'[^a-zA-Z0-9]', '', regex=+True)# reqex means reqular expression.
```

```python
customer_list['Phone_Number'].str.len()
```

```python
customer_list['Phone_Number']= customer_list['Phone_Number'].apply(lambda x :x[0:3] + '-' + x[3:6] + '-' + x[6:])
```

```python
customer_list
```

```python
customer_list['Phone_Number']= customer_list['Phone_Number'].str.replace('Na--', '')
customer_list['Phone_Number'] = customer_list['Phone_Number'].str.replace('nan--', '')
```

```python
customer_list
```

```python
customer_list[['Street Address', 'State', 'Zip Code']] = customer_list['Address'].str.split(',', n=2, expand=True)
```

```python
customer_list
```

```python
customer_list['Paying Customer'] = customer_list['Paying Customer'].str.replace('No', 'N')
customer_list['Paying Customer'] = customer_list['Paying Customer'].str.replace('Yes', 'Y')
customer_list['Do_Not_Contact'] = customer_list['Do_Not_Contact'].str.replace('No', 'Y')
customer_list['Do_Not_Contact'] = customer_list['Do_Not_Contact'].str.replace('Yes', 'Y')

customer_list.fillna('' , inplace=True)
```

```python
contact = customer_list[(customer_list['Do_Not_Contact']!= 'Y') & (customer_list['Phone_Number'].notnull())]
```

```python
contact.to_excel('Contact.xlsx')
```
