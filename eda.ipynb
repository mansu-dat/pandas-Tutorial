### Exploratory Data Analysis (EDA):


-Explore the distribution over time.

-Analyze the top- sellingproducts.

-Investigate the relationship between quantity sold and revenue.

-Identify any trends or patterns in customer behaviour 

```python
saless_over_time= df.groupby(df['Order Date'].dt.year)['Sales'].sum()# .dt.year is added to the code to get exact years
```

```python
saless_over_time
```

```python
plt.figure(figsize= (8, 4))
saless_over_time.plot(kind='line', color='skyblue', marker= 'o')
plt.title('Distribution of Sales over time')
plt.xlabel('Year')
plt.ylabel('Sales over Time')

plt.xticks(saless_over_time.index.astype(int))
plt.grid(True)
```

```python
top_sale_product= df.groupby('Product Name')['Sales'].sum().reset_index()
```

```python
top_sale_product
```

```python
top_sale_product=top_sale_product.sort_values(by='Sales',ascending = False)
top_n = 10
top_product =top_sale_product.head(top_n)
top_product
```

```python
plt.figure(figsize= (8, 4))
# top_product.plot(kind='bar',x='Sales', y='Product Name', color='skyblue')
top_product.plot(kind='bar', x='Product Name',y='Sales')
plt.title('top selling product')
plt.xlabel=('Sales')
plt.ylabel=('Global Sales(in Millions)')
plt.grid(True);


```

```python
top_product.plot(kind='hist', x='Product Name',y='Sales')
plt.title('top selling product')
plt.xlabel=('Sales')
plt.ylabel=('Global Sales(in millions)')
plt.grid(True);
```

```python
#As the dataset didn't include the revenue column. I will have to calculate it
df['Revenue'] = df['Quantity'] * df['Sales']
```

```python
#Investigate the relationship between quantity sold and revenue.
plt.figure(figsize=(14,6))
plt.scatter(df['Quantity'], df['Revenue'], alpha=0.5)
plt.title("Relationship Between Quantity Sold and Revenue")
plt.xlabel('Quantity Sold')
plt.ylabelRevenue
plt.gridTrue
plt.show
```

```python
#Group by discount level and calculate average revenue and quantity sold
discount_analysis = df.groupby('Discount').agg({
    'Revenue': 'mean',
    'Quantity': 'sum'
}).reset_index()
discount_analysis
```

```python
# Plot the impact of different discount levels on average revenue
plt.figure(figsize=(10, 6))
plt.bar(discount_analysis['Discount'], discount_analysis['Revenue'], color='skyblue')
plt.title('Discount Analysis: Impact on Average Revenue')
plt.xlabel('Discount Level')
plt.ylabel('Average Revenue per Transaction')
plt.grid(axis='y')
plt.show()
```
