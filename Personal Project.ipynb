```python
citatn_by_publictn= world_uni_ranking_df.groupby('citations')['publications'].value_counts().reset_index()
```

### Questions 2: what are the key factors influencing universities in different scores
#### load the dataset and filters by score
#### Group by score and calculate average citations 4 key factors
#### compare these averages accross scores


```python
ave_scores =world_uni_ranking_df.groupby('score').agg({'national_rank':'mean', 'quality_of_education':'mean', 
'quality_of_faculty':'mean', 'citations':'mean'}).reset_index()

```

```python
print(ave_scores)
```

```python
ave_scores.plot(x='score', kind='hist', figsize=(8,4))
plt.title('average score by citations');
```
