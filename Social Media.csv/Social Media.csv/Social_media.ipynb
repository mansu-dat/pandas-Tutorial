### The Clumns of the data sets is lengthy and have to reduce to minimal texts as shown below;

```python
social_media_df.columns = social_media_df.columns.str.replace('Do you use social media for professional purposes (e.g. LinkedIn, Twitter)?','Do use SM Professionally?')
social_media_df.columns = social_media_df.columns.str.replace('Have you ever experienced Cyberbullying or harassment on social media?','have you been bullied/harrased on SM?')
social_media_df.columns = social_media_df.columns.str.replace('How often do you adjust your privacy settings on social media?','Do you adjust private setting on SM')
social_media_df.columns = social_media_df.columns.str.replace('Have you purchased based on recommdtn on SM?\n','Have you purchased based on recommdtn on SM?')
social_media_df.columns = social_media_df.columns.str.replace('. Do you follow any Influencers or public figures on social media?\n','Do you follow influencer on SM?')
social_media_df.columns = social_media_df.columns.str.replace('What is your primary reason for using social media','Purpose of using SM')
social_media_df.columns = social_media_df.columns.str.replace('. Do you feel that social media has a positive or negative impact on your mental health?','Is there (+,-) impact on SM Mentally?')
social_media_df.columns = social_media_df.columns.str.replace('On average, how much time do you spend on social media each day?','Daily Timing on SM')
social_media_df.columns = social_media_df.columns.str.replace('. Which social media platform do you use the most?','Most Use SM')
social_media_df.columns = social_media_df.columns.str.replace('How often do you use social media platforms (e.g, Facebook, Instagram, Twitter, etc)','Time Spend on SM')
social_media_df.columns = social_media_df.columns.str.replace('Would you consider taking a break or permanently deleting your social media accounts?\n','Would you take break/permanently delete ur SM?')
```

```python
social_media_df
```

#### The row = 'Sex' at index 10 is = Male;Female and incorrect, so there is need to replace it with 'Male' and this shown below; 

```python
social_media_df['Sex'] = social_media_df['Sex'].replace('Male;Female','Male')
```

#### in the .describe above only count, unique, top, freq are shown instead of ( mean, median, mode, standard deviation), this is because most of the datatypes in the datasets are objects

```python
top_social_media= social_media_df.groupby('Most Use SM')['Sex'].value_counts().reset_index()
```

```python
top_social_media
```

```python
top_social_media=top_social_media.sort_values(by='Sex',ascending = False)
top_n = 10
top_social_media =top_social_media.head(top_n)
top_social_media
```

```python
plt.figure(figsize= (8, 4))
top_social_media.plot(kind='bar', x='Most Use SM',y='count')
plt.title('topsm_age')
plt.xlabel=('Most Use SM')
plt.ylabel=('count')
            
plt.grid(True);
```

### from the bar chart plot above, it shows that the most use social media is instagram and whatsapp

```python
gender_timing= social_media_df.groupby('Time Spend on SM')['Sex'].value_counts().reset_index()
```

```python
gender_timing
```

```python
plt.figure(figsize= (8, 4))
gender_timing.plot(kind='bar', x='Time Spend on SM',y='count')
plt.title('Gender Timing On Social Media')
plt.xlabel=('Time Spend on SM')
plt.ylabel=('count')            
plt.grid(True);
```

#### from this 2nd bar chart those that spend 'multiple time a day' on SM are the higest while the least are'once a day'
#### the males spend more times daily on SM than the females from this survey

```python
Use_socialmedia= social_media_df.groupby('Purpose of using SM?')['Do use SM Professionally?'].value_counts().reset_index()
```

```python
Use_socialmedia
```

```python
Use_socialmedia=Use_socialmedia.sort_values(by='Do use SM Professionally?',ascending = False)
n = 10
Use_socialmedia =Use_socialmedia.head(n)
Use_socialmedia
```

```python
plt.figure(figsize= (8, 4))
Use_socialmedia.plot(kind='box',x='Do use SM Professionally?',y='count')
plt.title('Purpose of Using Social Media')
plt.xlabel=('Do use SM Professionally?')
plt.ylabel=('count')            
plt.grid(True);
```

```python
bullying_sm= social_media_df.groupby('have you been bullied/harrased on SM?')['Age'].value_counts().reset_index()
```

```python
plt.figure(figsize= (8, 4))
bullying_sm.plot(kind='hist', x='Age',y='count')
plt.title('Purpose of Using Social Media')
plt.xlabel=('Age')
plt.ylabel=('count')            
plt.grid(True);
```
